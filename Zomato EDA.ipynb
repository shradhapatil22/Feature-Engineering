```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

## Conclusions from ratings
1. When rating is between 4.5 to 4.9 it is excellent
2. When rating is between 4.0 to 4.4 it is very good


matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green'])
```

# Observation
1. Not rated count is very high
2. Max ratings are between 2.5 to 3.4

# Count Plot

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green'])
```
