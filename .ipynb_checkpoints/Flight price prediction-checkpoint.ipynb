```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('FlightPriceTrain.xlsx')
```

```python
train_df.head()
```

```python
test_df=pd.read_excel('FlightPriceTest.xlsx')
```

```python
test_df.head()
```

```python
df = pd.concat([train_df,test_df], ignore_index=True)
df.head()
```

# Extract date month and year from date of journey

```python
df['Date_of_Journey']=pd.to_datetime(df['Date_of_Journey'])
```

```python
df['Day']=df['Date_of_Journey'].dt.day
df['Month']=df['Date_of_Journey'].dt.month
df['Year']=df['Date_of_Journey'].dt.year
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.columns
```

# Extracting arrival time from arrival time with dates

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0] 
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
```

```python
df['Arrival_Minute']=df['Arrival_Time'].str.split(':').str[1]
df['Arrival_Minute']=df['Arrival_Minute'].astype(int)
```

```python
df.head()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

# Extracting departure time from departure time with dates

```python
df['Dep_Time']=df['Dep_Time'].str.split(' ').str[0]
df['Dep_Hour']=df['Dep_Time'].str.split(':').str[0] 
df['Dep_Hour']=df['Dep_Hour'].astype(int)
df['Dep_Minute']=df['Dep_Time'].str.split(':').str[1]
df['Dep_Minute']=df['Dep_Minute'].astype(int)
df.drop('Dep_Time',axis=1,inplace=True)
df.info()
```

# Extracting number of stops and eliminating route column

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, 'nan':1, '4 stops':4})
```

```python
df.drop('Route',axis=1,inplace=True)
```

# Making duration into minutes 

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df.drop('Duration',axis=1,inplace=True)
```

# Label Encoding for categorical values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelEncoder=LabelEncoder()
```

```python
encoding_cols=['Airline','Source','Destination','Additional_Info']
label_encoders ={}
for column in encoding_cols:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
   
    
df.info()
```
